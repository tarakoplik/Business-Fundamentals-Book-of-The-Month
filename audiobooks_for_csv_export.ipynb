```python
#1790 was the only column loaded in incorrectly let's fix that first
df.loc[1790,'title'] = 'Her Last Wish (A Rachel Gift Mystery--Book 1)'
df.loc[1790,'author'] = 'Narrated By: Blake Pierce'
df.loc[1790,'price'] = '$1.99'
df.loc[1790,'date'] = 'Date: November 2021'
df.loc[1790,'narrator'] = 'Mary Sarah'
df.loc[1790,'genre'] = 'Mystery, Thriller & Horror > Detective Stories'
df.loc[1790,'duration'] = 'Duration: 6 hours 18 minutes'
df.loc[1790,'rating'] = 3.98
```

```python
df.loc[1790]
```

## Author

```python
updated_author = []
for auth in list(df['author']):
    auth = auth[12:]
    updated_author.append(auth)
df['author'] = updated_author
```

## Price

```python
updated_price = []
for p in list(df['price']):
    start = (p.find("$")) + 1 
    end = (p.find(".")) + 3
    new_price = p[start:end]
    updated_price.append(new_price)
df['price'] = updated_price
```

## Date

```python
new_date = []
date_list = list(df['date'])
for d in date_list:
    d = d[6:]
    d = d.strip()
    new_date.append(d)
    
df['date'] = new_date
```

```python
from datetime import datetime as dt

df['date'] = pd.to_datetime(df.date)
```

```python
import datetime
date_list = list(df['date'])

comp_date = datetime.datetime(2022,6,11,00,00,00)
date_list = list(df['date'])
pre_order = []
for i in date_list:
    if i > comp_date:
        pre_status = "yes"
    else:
        pre_status = "no"
    pre_order.append(pre_status)
    
df['is_pre_order'] = pre_order
```

## Narrator

```python
new_nar = []
for n in list(df['narrator']):
    n = n[12:]
    n = n.strip()
    new_nar.append(n)
df['narrator'] = new_nar
```

## Genres

```python
genres = list(df['genre'])
genre_1 = []
genre_2 = []
for g in genres:
    split_point = g.find(">")
    g_1 = g[:split_point]
    g_1 = g_1.strip()
    genre_1.append(g_1)
    g_2 = g[(split_point+1):]
    g_2 = g_2.strip()
    genre_2.append(g_2)
    
df['genre_1'] = genre_1
df['genre_2'] = genre_2
```

```python
df.drop(columns = ['genre'],inplace=True)
```

## Rating

```python
updated_rating = []
rate_list = list(df['rating'])
for r in rate_list:
    if r == 'not_found':
        new_r = 0
    else:
        new_r = r
    updated_rating.append(new_r)
df['rating'] = updated_rating
```

```python
column_names =['rank','title','author','price','narrator','rating','date','is_pre_order','genre_1','genre_2','duration','duration_mins']
df = df.reindex(columns=column_names)
```

# Export data to csv

```python
df.to_csv('audiobook_data.csv',index=False)
```
